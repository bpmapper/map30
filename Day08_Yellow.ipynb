## Plot placenames in Great Britain that contain the word "Sunny"

```python
# Import packages
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Download OS Zoomstack from https://osdatahub.os.uk/ and read in names layer
myzoom='C:\GIS\Zoomstack\OS_Open_Zoomstack.gpkg'
gdf=gpd.read_file(myzoom, layer = 'names')
gdf.sample(3)
```

```python
gdfG=gdf[gdf.name1.str.contains('Sunny')]
gdfG
```

```python
gdfG.name1.value_counts()[:10]
```

```python
gdfGB=gpd.read_file(r'C:\GIS\Borders\Data\GB\high_water_polyline.shp')
```

```python
fig, ax = plt.subplots(figsize=(15,15))
gdfGB.plot(ax=ax, color='white')
gdfG.plot(ax=ax, color='yellow')
ax.text(.02,.98,'Day 08 - Yellow',transform=ax.transAxes, fontsize=15).set_color('white')
ax.text(.02,.96,'GB Placenames that contain the word "Sunny"',transform=ax.transAxes, fontsize=15).set_color('white')       
ax.text(.02,.02,'Data: Names layer in OS Zoomstack geopackage from osdatahub.os.uk',transform=ax.transAxes, fontsize=10).set_color('white')       
ax.text(.02,.01,'Code: Python/Jupyter Notebook at github.com/bpmapper/map30',transform=ax.transAxes, fontsize=10).set_color('white')       
ax.set_facecolor('black')
plt.tick_params(axis='both',which='both',bottom=False,left=False,labelbottom=False,labelleft=False)
plt.show()
fig.savefig('Day08_Yellow.png')
```
