# Day 4 - Hexagons

## Plot placenames in Great Britain that contain the letter X

```python
# Import packages
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Download OS Zoomstack from https://osdatahub.os.uk/ and read in names layer
myzoom='C:\GIS\Zoomstack\OS_Open_Zoomstack.gpkg'
gdf=gpd.read_file(myzoom, layer = 'names')
gdf.sample(3)
```

```python
# most name languages not classified - check impact of Welsh points e.g. Gogledd for North
gdf[gdf.name1.str.contains('x')]
```

```python
fig, ax=plt.subplots(figsize=(10,10))
#gdf.plot(ax=ax)
ax.hexbin(gdf.geometry.x, gdf.geometry.y,gridsize=30, cmap='YlOrBr', alpha=1)

ax.text(.02,.95,'Day 04 - HeXagons',transform=ax.transAxes, fontsize=15).set_color('black')
ax.text(.02,.90,'GB Placenames that contain the letter X',transform=ax.transAxes, fontsize=15).set_color('black')       
ax.text(.02,.05,'Data: Names layer in OS Zoomstack geopackage from osdatahub.os.uk',transform=ax.transAxes, fontsize=10).set_color('black')       
ax.text(.02,.01,'Code: Python/Jupyter Notebook at github.com/bpmapper/map30',transform=ax.transAxes, fontsize=10).set_color('black')       

plt.tick_params(axis='both',which='both',bottom=False,left=False,labelbottom=False,labelleft=False)
plt.show()
fig.savefig('Day04_Hexagons.png')
```
